```python
stack1 =  [
          ['Z','N'],
          ['M','C','D'],
          ['P']
          ]


moves = [[1,2,1],[3,1,3],[2,2,1],[1,1,2]]

def stack_after_mov(moves,stack1):
  for play in moves:
    number_of_pieces = play[0]
    from_stack = play[1] - 1
    to_stack   = play[2] - 1

    size_from = len(stack1[from_stack])-1
    new_val = []
    for i in range(0,number_of_pieces):
      new_val.append(stack1[from_stack].pop(size_from-i))

    for i in range(0,len(new_val)):
      stack1[to_stack].append(new_val[i])
  return stack1

new_stack = stack_after_mov(moves,stack1)
```
